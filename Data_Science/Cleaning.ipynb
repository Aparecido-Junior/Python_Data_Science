# Handling Missing Data

```python
df.dropna()
df.dropna(how='all')
df.dropna(thresh=2)
df.dropna(subset=['City'])
```

```python
df[df['City'].notna()]
```

```python
df[df.isna().any(axis=1)].dropna()
df[df.isna().any(axis=1)].dropna(subset=['Student','Class'])
df = df.dropna(subset=['Student','Class'], inplace=True)
df
```

# IMPUTING MISSING DATA

```python
df.Grade
df[df.Grade.isna()]
```
